## Tampilkan dataset

```python
import pandas as pd

# Baca file CSV
df = pd.read_csv('datasetUMPOHoax.csv')

# Tampilkan beberapa baris
df.head()
```

## Filter label Hoax dan ambil kolom

```python
# Ambil data berlabel 'hoax' (case-insensitive)
df_hoax = df[df['label'].str.lower() == 'hoax']

# Ambil hanya kolom tweet
df_hoax = df_hoax[['tweet']].dropna()
df_hoax = df_hoax[df_hoax['tweet'].str.strip() != '']
```

## Bersihkan teks

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)  # hapus URL
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)  # hapus karakter selain huruf/angka/spasi
    text = re.sub(r"\s+", " ", text)  # hapus spasi ganda
    return text.strip()

# Bersihkan semua tweet
df_hoax['cleaned'] = df_hoax['tweet'].apply(clean_text)
```

```python
# Tambahkan token khusus
df_hoax['input_text'] = '<start> ' + df_hoax['tweet'] + ' <end>'
```

## Tokenisasi dan sequence

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Tokenizer tanpa batasan num_words
tokenizer = Tokenizer(oov_token='<OOV>')
tokenizer.fit_on_texts(df_hoax['cleaned'])

total_words = len(tokenizer.word_index) + 1

# Buat input sequences (n-gram)
input_sequences = []
for line in df_hoax['cleaned']:
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_seq = token_list[:i+1]
        input_sequences.append(n_gram_seq)

# Padding sequences
max_seq_len = max([len(seq) for seq in input_sequences])
input_sequences = pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre')
```

```python
import numpy as np
from keras.utils import to_categorical

X = input_sequences[:, :-1]
y = input_sequences[:, -1]
y = to_categorical(y, num_classes=total_words)
```

## Train model LSTM

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional

model = Sequential()
model.add(Embedding(input_dim=total_words, output_dim=100, input_length=max_seq_len-1))
model.add(Bidirectional(LSTM(150)))
model.add(Dropout(0.2))
model.add(Dense(total_words, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
# Latih model
history = model.fit(X, y, epochs=25, verbose=1, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt

# Visualisasi Akurasi
plt.plot(history.history['accuracy'], label='Akurasi Training')
plt.plot(history.history['val_accuracy'], label='Akurasi Validasi')
plt.title('Akurasi Model Selama Training')
plt.xlabel('Epoch')
plt.ylabel('Akurasi')
plt.legend()
plt.grid(True)
plt.show()

# Visualisasi Loss
plt.plot(history.history['loss'], label='Loss Training')
plt.plot(history.history['val_loss'], label='Loss Validasi')
plt.title('Loss Model Selama Training')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

## Fungsi generate

```python
import random

def sample_word(preds, temperature=1.0):
    preds = np.asarray(preds).astype('float64')
    preds = np.log(preds + 1e-10) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    probas = np.random.multinomial(1, preds, 1)
    return np.argmax(probas)
```

```python
def generate_hoax_from_word(seed_word, tokenizer, model, max_seq_len, max_words=10, temperature=1.0):
    seed_text = seed_word.lower()
    output_text = seed_text
    for _ in range(random.randint(6, max_words)):
        token_list = tokenizer.texts_to_sequences([output_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_seq_len-1, padding='pre')
        preds = model.predict(token_list, verbose=0)[0]
        predicted_index = sample_word(preds, temperature)
        predicted_word = tokenizer.index_word.get(predicted_index, '')
        if predicted_word == '' or predicted_word == '<OOV>':
            break
        output_text += ' ' + predicted_word
    return output_text
```

```python
def extract_spok(text):
    doc = nlp(text)
    spok = []
    for sent in doc.sents:
        subj, pred, obj, keterangan = None, None, None, []
        for word in sent:
            if word.dep_ in ['nsubj', 'nsubj:pass']:
                subj = word.text
            elif word.pos_ == 'VERB':
                pred = word.text
            elif word.dep_ in ['obj', 'iobj']:
                obj = word.text
            elif word.dep_ == 'obl':
                keterangan.append(word.text)
        spok.append({
            'kalimat': sent.text,
            'subjek': subj,
            'predikat': pred,
            'objek': obj,
            'keterangan': " ".join(keterangan)
        })
    return spok
```
